## Basic Plottings

#### Line Plot

```python
# Line Plot
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

```python
plt.plot(x, y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("A Simple Line Chart")
plt.show()
```

#### Scatter Plot

```python
# Scatter Plot
plt.scatter(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("A Simple Scatter Plot")
plt.show()
```

#### Bar Plot

- Used to compare Quantities

```python
# Bar Plot
Categories = ['A', 'B', 'C', 'D']
Values = [3, 5, 7, 11]
```

```python
plt.bar(Categories, Values)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("A Simple Bar Plot")
plt.show()
```

#### Histograms
- Used to show the distribution of our dateset.

```python
data = [1,2,3,4,4,5,5,5,6,7,7,7,8,8,8,8,9,9]
plt.hist(data, bins = 5)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
data = [1,2,3,4,4,5,5,5,6,7,7,7,8,8,8,8,9,9]
plt.hist(data, bins=6)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

- Histogram is a graph showing frequency distributions.
- It is a graph showing the number of observations within each given interval

- **bin**: width of each bin = (max value of data - min value of data)/total number of bins
- The height of each bion shows how many values from that data fall into that range.

#### Box Plot
- Show the distribution of the data based on 5 summary statistics: 
    - Minimum, First Quartile, Median and Third Quartile, Maximum
- Helps to find the Outliers.

```python
plt.boxplot(data)
plt. title('Box Plot')
plt.show
```

```python
data1 = [1,6,6,7,8,8,8,8,8,9]
plt.boxplot(data1)
plt.title('Box Plot with outlier')
plt.show
```

### Basic Plotting

#### Scatter Plot

```python
data = {
    'Size': [1500, 2500, 1800, 2200, 2400],
    'No of Bedrooms': [3,4,3,5,4],
    'No of Bathrooms': [2,3, 2, 3,2],
    'Year Built': [2000, 2010, 2005, 2021, 2018],
    'Location': ['Suburb', 'City', 'City', 'Suburb', 'Rural'],
    'Price': [300000, 450000, 350000, 500000, 420000]
}
df = pd.DataFrame(data)
df
```

```python
# Scatter Plot
sns.scatterplot(x = 'No of Bedrooms', y = 'No of Bathrooms', data=df)
plt.title('Scatter Plot to Bedroom and Bathroom Number')
plt.show()
```

#### Line Plot

```python
# Line Plot
sns.lineplot(x='Year Built', y="No of Bedrooms", data=df)
plt.title('Line Plot')
plt.show()
```

#### Bar Plot  

```python
sns.barplot(x='Location', y='Price', data=df)
plt.title('Bar Plot for Price per the Location')
plt.show()
```
